```python
import pandas as pd
tdf = pd.read_csv("titanic.csv")
```

```python
tdf.info()
```

```python
tdf.isnull().sum()
```

```python
tdf['PassengerId'] = tdf['PassengerId'].astype(object)
tdf['Survived'] = tdf['Survived'].astype(bool)
tdf['Pclass'] = pd.Categorical(tdf['Pclass'])
tdf['Sex'] = pd.Categorical(tdf['Sex'])
tdf['Embarked'] = pd.Categorical(tdf['Embarked'])
```

```python
tdf.info()
```

```python
tdf.describe()
```

```python
print(tdf['Age'].median())
print(tdf['SibSp'].median())
print(tdf['Parch'].median())
print(tdf['Fare'].median())

print(tdf['Age'].mode())
print(tdf['SibSp'].mode())
print(tdf['Parch'].mode())
print(tdf['Fare'].mode())
```

```python
import matplotlib.pyplot as plt
path = "C:\\Users\\Jane\\Documents\\PDD-DA@Langara\\Spring Semester 1.0\\CPSC 4800\\Assignments\\Assignment 3\\Assignment_3_2_Plots\\"
plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.hist(tdf['Age'], color='skyblue', edgecolor='black')
plt.savefig(path + "Age_histogram.png")
```

```python
plt.title("SibSp Histogram")
plt.xlabel("SibSp")
plt.ylabel("Frequency")
plt.hist(tdf['SibSp'], color='skyblue', edgecolor='black')
plt.savefig(path + "SibSp_histogram.png")
```

```python
plt.title("Parch Histogram")
plt.xlabel("Parch")
plt.ylabel("Frequency")
plt.hist(tdf['Parch'], color='skyblue', edgecolor='black')
plt.savefig(path + "Parch_histogram.png")
```

```python
plt.title("Fare Histogram")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.hist(tdf['Fare'], color='skyblue', edgecolor='black')
plt.savefig(path + "Fare_histogram.png")
```

```python
surv = tdf['Survived'].value_counts()
pc = tdf['Pclass']. value_counts(sort = False)
sx = tdf['Sex'].value_counts()
emb = tdf['Embarked'].value_counts()
print(surv, pc, sx, emb)
```

```python
plt.bar(["0","1"], surv , color = "skyblue", edgecolor='black')
plt.title("Survived")
plt.xlabel('Survived/Not')
plt.ylabel('Frequency')
plt.savefig(path + "Survived_barplot.png")
```

```python
pck_lst = pc.keys().tolist()
pc_labels = [str(x) for x in pck_lst]
plt.bar(pc_labels, pc, color = "skyblue",  edgecolor='black')
plt.title("Pclass")
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.savefig(path + "Pclass_barplot.png")
```

```python
plt.bar(["Male", "Female"], sx, color = "skyblue",  edgecolor='black')
plt.title("Sex")
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.savefig(path + "Sex_barplot.png")
```

```python
plt.bar(emb.keys() ,emb, color = "skyblue",  edgecolor='black')
plt.title("Port of Embarkation")
plt.xlabel('Ports')
plt.ylabel('Frequency')
plt.savefig(path + "Emb_barplot.png")
```

```python
#finding overall survival rate
true_counts = tdf[tdf.Survived == 1].shape[0]
osr = true_counts/len(tdf['Survived'])
print(osr)
```

```python
pclass_table = tdf.groupby('Pclass', observed = False)['Survived']
print(pclass_table.mean())
```

```python
pclass_table_c = pclass_table.value_counts().unstack()
pclass_table_c.plot(kind = "bar", stacked = True)
plt.savefig(path + "pclass_survived.png")
```

```python
sex_table = tdf.groupby('Sex', observed = False)['Survived']
sex_table.mean()
```

```python
sex_table_c = sex_table.value_counts().unstack()
sex_table_c.plot(kind = "bar", stacked = True)
plt.savefig(path + "sex_survived.png")
```

```python
import numpy as np
new_tdf = tdf.dropna()
corr_coeff= np.corrcoef(new_tdf["Age"], new_tdf["Survived"])
print(corr_coeff)
```
